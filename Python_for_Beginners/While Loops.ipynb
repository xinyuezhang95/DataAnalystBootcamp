## While Loops

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
# break statement
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
# else statement
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
# continue statement
number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
else:
    print('No longer < 5')

# The result of running this is that it will continue printing 3 since the +1 is after 'continue'
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')
```
